```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
df= pd.read_csv(r"C:\Users\Admin\Desktop\projects\E-commerce project\archive\online_retail.csv")
pd.set_option('display.max_columns', None)
df
```

```python
# Remove all duplicate rows
df = df.drop_duplicates()
df
```

```python
df.head()
df.info()
df.describe()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df
```

```python
# Remove Cancelled orders (Orders where the InvoiceNo starts with 'c')
# Remove orders with missing CustomerID
df = df[~df['InvoiceNo'].str.startswith('C')].copy()
df = df.dropna(subset=['CustomerID']).copy()
df
```

```python
# Add a new column, Total price per row (Quantity * Unit Price)
df['TotalPrice'] = df['Quantity'] * df['UnitPrice'].round(2)
df
```

```python
# Sales distribution by country
country_sales= (
    df.groupby('Country')['TotalPrice']
        .sum()
        .sort_values(ascending=False)
        )
country_sales.head(10).plot(kind='bar', title='Top 10 Countries by Revenue')
plt.show()
```

```python
# Top 10 products that generate the most revenue
product_sales = (
    df.groupby("Description")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
)
product_sales.head(10).plot(kind="barh")
```

```python
# Top 10 highest spending customers
customer_spending = (
    df.groupby("CustomerID")["TotalPrice"]
    .sum()
    .sort_values(ascending=False)
)
customer_spending.head(10).plot(kind="barh")
```

```python
product_sales.head(50).to_csv(r"C:\Users\Admin\Desktop\projects\E-commerce project\top_products.csv")
country_sales.to_csv(r"C:\Users\Admin\Desktop\projects\E-commerce project\country_sales.csv")
customer_spending.to_csv(r"C:\Users\Admin\Desktop\projects\E-commerce project\customer_spending.csv")
df.to_csv(r"C:\Users\Admin\Desktop\projects\E-commerce project\online_retail_new.csv", index=False)
```
